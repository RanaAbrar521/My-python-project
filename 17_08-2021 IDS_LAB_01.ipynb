```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('olympics.csv', index_col = 0, skiprows=1)
df.head()
```

```python
for i in df.columns:
    if i[:2] == "01":
        df.rename(columns = {i:'Gold' + i[4:]},inplace = True)
    if i[:2] == "02":
        df.rename(columns = {i:'Silver' + i[4:]},inplace = True)
    if i[:2] == "03":
        df.rename(columns = {i:'Bronze' + i[4:]},inplace = True)
    if i[:1] == "№":
        df.rename(columns = {i:'#' + i[1:]},inplace = True)
        
names_ids = df.index.str.split('\s\(')
df.index = names_ids.str[0]
df['ID'] = names_ids.str[1].str[:3]
```

```python
df.drop('Totals', inplace = True)
def answer_two():
    return (df['Gold'] - df['Gold.1']).idxmax()
answer_two()
```

```python
def answer_three():
    filter1  = df[( df['Gold']>0) & (df['Gold.1']>0)]
    m = (filter1['Gold'] - filter1['Gold.1'])/filter1['Gold.2']
    return m.idxmax()
answer_three()
```
