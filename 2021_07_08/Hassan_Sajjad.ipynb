```python
# Question # 01
import datetime as dt
import time as tm

def get_date(days):
    if (days % 2) != 0:
        return dt.datetime.today() + dt.timedelta(days=days)
    else:
        return dt.datetime.today() - dt.timedelta(days=days)

daysODD = 9
daysEVEN = 6
print(f"EVEN Date : {get_date(daysEVEN)}")
print(f"ODD Date : {get_date(daysODD)}")
```
