```python
class Student:
    def __init__(self,name,batch,address,age,father_name,city):
        self.name=name
        self.batch = batch
        self.address = address
        self.age = age
        self.father_name = father_name
        self.city = city
    
    def set_name(self,name):
        self.name=name
        
    def set_batch(self,batch): 
        self.batch=batch
        
    def set_address(self,address):
        self.address=address
        
    
    def get_name(self):
        return self.name
    
    def get_batch(self):
        return self.batch
    
    def get_address(self):
        return self.address

class DevnationStudents(Student):
    def __init__(self,name,batch,address,age,father_name,city,courses=[],marks=[]):
        super().__init__(name,batch,address,age,father_name,city)
        
    
    def set_add_course(self,add_course):

        for i in add_course:
            if i not in self.add_course:
                self.add_course.append(i)

    
    
    def set_add_course_marks(self,add_course_marks):
        
        self.add_course_marks= add_course_marks
        
        
    def get_add_course(self):
        return self.add_course
    
    def get_add_course_marks(self):
        return self.add_course_marks
```

```python
def backward_number(num):
    Number = int(input("Please Enter any Number: "))
    Reverse = 0
    while(Number > 0):
        Reminder = Number %10
        Reverse = (Reverse *10) + Reminder
        Number = Number //10

    print("\n Reverse of entered number is = %d" %Reverse)
```

```python
#Q10
#function = remove dublication
```

```python
def remove_dublication():
    my_list = [1, 2, 3, 1, 2, 4, 5, 4 ,6, 2]
    print("List Before ", my_list)
    temp_list = []

    for i in my_list:
        if i not in temp_list:
            temp_list.append(i)

    my_list = temp_list

    print("List After removing duplicates ", my_list)
```

```python
remove_dublication()
```
