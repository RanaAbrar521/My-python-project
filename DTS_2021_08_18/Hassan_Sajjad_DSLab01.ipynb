```python
def answer_six():
    return dfc[dfc['SUMLEV'] == 50].groupby('STNAME')['CENSUS2010POP'].nlargest(3).groupby('STNAME').sum().nlargest(3).index.values.tolist()
```

```python
answer_six()
```

```python
def answer_seven():
    df=dfc[dfc['SUMLEV'] == 50]
    columns_to_keep = ['CTYNAME', 'POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012', 'POPESTIMATE2013','POPESTIMATE2014', 'POPESTIMATE2015']
    df = df[columns_to_keep]
    df = df.set_index('CTYNAME')
    df['min'] = (df.min(axis=1, skipna=True))
    df['max'] = (df.max(axis=1, skipna=True))
    df['diff'] = df['max'] - df['min']
    cty = ((df['diff']).idxmax())
    
    return cty
answer_seven()
```

```python
def answer_eight():
    counties = dfc[dfc['SUMLEV'] == 50]
    region = counties[(counties['REGION'] == 1) | (counties['REGION'] == 2)]
    washington = region[region['CTYNAME'].str.startswith("Washington")]
    grew = washington[washington['POPESTIMATE2015'] > washington['POPESTIMATE2014']]
    return grew[['STNAME', 'CTYNAME']]

answer_eight()
```
