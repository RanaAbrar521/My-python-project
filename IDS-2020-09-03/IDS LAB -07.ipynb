```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("pulls.csv")
df1 = pd.read_csv("pull_files.csv")
```

```python
def Question_1():
    display(df.head())
    display(df1.head())
Question_1()
```

```python
def Question_2():
    df['UTS'] = pd.to_datetime(df['date'],utc=True)
    return df.head()
Question_2()
```

```python
df_merged = pd.merge(df,df1,how='inner',on='pid')
def Question_3():
    return df_merged.head()
Question_3()
```

```python
def Question_5():
    pulls_to_user = df_merged.groupby('user')['pid'].count()
    pulls_to_user.plot(kind='hist',figsize=(12,6),grid=True)
    plt.title('Contribution of Users')
    plt.xlabel('Users')
    plt.ylabel('Pull Requests')
    plt.show()
Question_5()    
```

```python
def Question_6():
    return df_merged.nlargest(20,'UTS')
Question_6()
```

```python
def Question_7():
    file = 'src/compiler/scala/reflect/reify/phases/Calculate.scala'
    file_df = df_merged[df_merged['file'] == file]
    result = file_df.groupby('user').count()
    return result.sort_values(['pid'],ascending = False).nlargest(3,'pid')
Question_7()
```

```python
def Questio_8():    
    file = 'src/compiler/scala/reflect/reify/phases/Calculate.scala'
    file_df = df_merged[df_merged['file'] == file]
    result = file_df.nlargest(10,'UTS')
    return t(result['user'].unique())
Questio_8()
```
