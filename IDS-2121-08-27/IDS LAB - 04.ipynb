```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tv.csv')
df1 = pd.read_csv('super_bowls.csv')
df2 = pd.read_csv('halftime_musicians.csv')
```

```python
def Question_1():
    display(df.head())
    display(df1.head())
    display(df2.head())
Question_1()
```

```python
def question_2():
    df.info()
    print('\n')
    df2.info()
question_2()
```

```python
def Question_3():
    df1.plot(kind='line',x='super_bowl',y='combined_pts',figsize=(14,7),grid=True,color='black')
    font1 = {'family':'serif','color':'blue','size':20}
    font2 = {'family':'serif','color':'darkred','size':15}
    plt.title('Combined Ball Distribution-Line Graph',loc='left',fontdict=font1)
    plt.ylabel('combined_pts',fontdict=font2)
    plt.xlabel('super_bowl',fontdict=font2)
    plt.show()
    df1.plot(kind='bar',x='super_bowl',y='combined_pts',figsize=(14,7),grid=True,color='black')
    font1 = {'family':'serif','color':'blue','size':20}
    font2 = {'family':'serif','color':'darkred','size':15}
    plt.title('Combined Ball Distribution-Bar Graph',loc='left',fontdict=font1)
    plt.ylabel('combined_pts',fontdict=font2)
    plt.xlabel('super_bowl',fontdict=font2)
    plt.show()
    display(df1[df1['combined_pts']==df1['combined_pts'].max()])
    display(df1[df1['combined_pts']==df1['combined_pts'].min()])
Question_3()
```

```python
def Question_4():
    df1.plot(kind='line',x='super_bowl',y='difference_pts',figsize=(14,7),grid=True,color='green')
    font1 = {'family':'serif','color':'blue','size':20}
    font2 = {'family':'serif','color':'darkred','size':15}
    plt.title('Combined Ball Distribution-Line Graph',loc='left',fontdict=font1)
    plt.ylabel('combined_pts',fontdict=font2)
    plt.xlabel('super_bowl',fontdict=font2)
    plt.show()
    df1.plot(kind='bar',x='super_bowl',y='difference_pts',figsize=(14,7),grid=True,color='green')
    font1 = {'family':'serif','color':'blue','size':20}
    font2 = {'family':'serif','color':'darkred','size':15}
    plt.title('Combined Ball Distribution-Bar Graph',loc='left',fontdict=font1)
    plt.ylabel('combined_pts',fontdict=font2)
    plt.xlabel('super_bowl',fontdict=font2)
    plt.show()
Question_4()
```

```python
def Question_5():
    data = pd.merge(df,df1,on='super_bowl',how='outer')
    return sns.regplot(x='difference_pts',y='share_household',data=data,color='brown')
Question_5()
```

```python
def Question_6():
    df.plot(kind='line',x='super_bowl',y='avg_us_viewers',grid=True,color='r')
    df.plot(kind='line',x='super_bowl',y='rating_household',grid=True,color='r')
    df.plot(kind='line',x='super_bowl',y='ad_cost',grid=True,color='r')
Question_6()
```
