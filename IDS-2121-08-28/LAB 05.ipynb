```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('coinmarketcap_06122017.csv')
df1=pd.read_csv('coinmarketcap_06012018.csv')
```

```python
def Question_2():
    return df[['id','market_cap_usd']].count()
Question_2()
```

```python
def Question_3():
    valid_df=df[df['market_cap_usd']>0]
    return valid_df[['id','market_cap_usd']].count()
Question_3()
```

```python
def Questio_4():
    top_ten=df.head(10)
    sum_of_all_market=top_ten[['market_cap_usd']].sum().iloc[0]
    top_ten['market_cap_pct']=(top_ten['market_cap_usd']/sum_of_all_market)*100
    top_ten.plot(kind='bar',x='id',y='market_cap_pct',figsize=(12,6),grid=True)
    font1 = {'family':'serif','color':'blue','size':20}
    font2 = {'family':'serif','color':'darkred','size':15}
    plt.title('Market Capitalization of Top 10 Coins',loc='left',fontdict=font1)
    plt.ylabel('market_cap_usd',fontdict=font2)
    plt.xlabel('id',fontdict=font2)
    plt.show()

Questio_4()
```

```python
def Questio_5():
    top_ten=df.head(10)
    top_ten.plot(kind='bar',x='id',y='market_cap_usd',figsize=(12,6),grid=True,log=True,color= ['r','green','r','k','k','pink','brown','r','orange','green']) 
    font1 = {'family':'serif','color':'blue','size':20}
    font2 = {'family':'serif','color':'darkred','size':15}
    plt.title('Market Capitalization of Top 10 Coins',loc='left',fontdict=font1)
    plt.ylabel('market_cap_usd',fontdict=font2)
    plt.xlabel('id',fontdict=font2)
    plt.show()
Questio_5()    
```

```python
def Questio_6():
    df.plot(kind='line',x='id',y='percent_change_24h',figsize=(12,6),grid=True) 
    font1 = {'family':'serif','color':'blue','size':20}
    font2 = {'family':'serif','color':'darkred','size':15}
    plt.title('Exploring Volatility',loc='left',fontdict=font1)
    plt.ylabel('percent_change_24h',fontdict=font2)
    plt.xlabel('id',fontdict=font2)
    plt.show()
    df.plot(kind='line',x='id',y='percent_change_7d',figsize=(12,6),grid=True) 
    font1 = {'family':'serif','color':'blue','size':20}
    font2 = {'family':'serif','color':'darkred','size':15}
    plt.title('Exploring Volatility',loc='left',fontdict=font1)
    plt.ylabel('percent_change_7d',fontdict=font2)
    plt.xlabel('id',fontdict=font2)
    plt.show()
Questio_6()    
```

```python
def Questio_7():
    top_ten=df.nlargest(10,'percent_change_24h')
    top_ten.plot(kind='bar',x='id',y='percent_change_24h',figsize=(12,6),grid=True) 
    font1 = {'family':'serif','color':'blue','size':20}
    font2 = {'family':'serif','color':'darkred','size':15}
    plt.title('Exploring Volatility',loc='left',fontdict=font1)
    plt.ylabel('percent_change_24h',fontdict=font2)
    plt.xlabel('id',fontdict=font2)
    plt.show()
    top_ten=df.nsmallest(10,'percent_change_24h')
    top_ten.plot(kind='bar',x='id',y='percent_change_7d',figsize=(12,6),grid=True) 
    font1 = {'family':'serif','color':'blue','size':20}
    font2 = {'family':'serif','color':'darkred','size':15}
    plt.title('Exploring Volatility',loc='left',fontdict=font1)
    plt.ylabel('percent_change_7d',fontdict=font2)
    plt.xlabel('id',fontdict=font2)
    plt.show()
Questio_7()   
```
