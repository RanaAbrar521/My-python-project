```python
import pandas as pd

df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)

for col in df.columns:
    if col[:2] == '01':
        df.rename(columns={col:'Gold' + col[4:]}, inplace = True)
    if col[:2] == '02':
        df.rename(columns={col:'Silver' + col[4:]}, inplace = True)
    if col[:2] == '03':
        df.rename(columns={col:'Bronze' + col[4:]}, inplace = True)
    if col[:1] == '№':
        df.rename(columns={col:'#' + col[1:]}, inplace = True)
        
names_ids = df.index.str.split('\s\(')

df.index = names_ids.str[0]
df['ID'] = names_ids.str[1].str[:3]

df = df.drop('Totals')
df.head(10)
```

```python
def answer_zero():
    return df.iloc[0]
```

```python
answer_zero()
```

```python
def answer_one():
    return df['Gold'].idxmax()
```

```python
def answer_two():
    mini = df['Gold'] - df['Gold.1']
    return mini.idxmax()
```

```python
def answer_three():
    filter1  = df[( df['Gold']>0) & (df['Gold.1']>0)]
    m = (filter1['Gold'] - filter1['Gold.1'])/filter1['Gold.2']
    return m.idxmax()
```

```python
answer_three()
```

```python
census_df = pd.read_csv('census.csv')
census_df.head(10)
```

```python
def answer_five():
    return (census_df['STNAME'].value_counts().idxmax())
```

```python
answer_five()
```

```python
def answer_six():
    populous = census_df.nlargest(3, 'CENSUS2010POP')
    return list(populous['STNAME'])
```

```python
answer_six()
```

```python
def answer_seven():
    #taking the list of the column names
    popest = ["POPESTIMATE2010","POPESTIMATE2011","POPESTIMATE2012","POPESTIMATE2013","POPESTIMATE2014","POPESTIMATE2015"]
    # subtracting the maximum with minimum and getting the highest index number and use iloc and return the answer
    return (census_df.iloc[(census_df[popest].max(axis=1) - census_df[popest].min(axis=1)).argmax()]['CTYNAME'])
    
```

```python
answer_seven()
```

```python
def answer_eight():
    filter_region = census_df[(census_df['REGION']==1) | (census_df['REGION']==2)]
    washington_named = filter_region[filter_region['CTYNAME'].str.startswith("Washington")]
    popest = washington_named[washington_named['POPESTIMATE2015'] > washington_named['POPESTIMATE2014']]
    return (popest[['STNAME','CTYNAME']])
```

```python
answer_eight()
```
