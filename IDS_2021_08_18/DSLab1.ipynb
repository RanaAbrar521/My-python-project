```python
import pandas as pd

df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)

for col in df.columns:
    if col[:2] == '01':
        df.rename(columns={col:'Gold' + col[4:]}, inplace = True)
    if col[:2] == '02':
        df.rename(columns={col:'Silver' + col[4:]}, inplace = True)
    if col[:2] == '03':
        df.rename(columns={col:'Bronze' + col[4:]}, inplace = True)
    if col[:1] == '№':
        df.rename(columns={col:'#' + col[1:]}, inplace = True)
        
names_ids = df.index.str.split('\s\(')

df.index = names_ids.str[0]
df['ID'] = names_ids.str[1].str[:3]

df = df.drop('Totals')
df.head(10)
```

```python
def answer_zero():
    return df.iloc[0]
answer_zero()
```

```python
def answer_one():
    return df['Gold'].idxmax()
answer_one()
```

```python
def answer_two():
    return (df['Gold'] - df['Gold.1']).idxmax()
answer_two()
```

```python
def answer_three():
    x = df[(df['Gold']>=1) & (df['Gold.1']>=1)]
    y = (x['Gold'] - x['Gold.1'])/x['Gold.2']
    return y.idxmax()
answer_three()
```

```python
def answer_four():
    points = df['total_points'] = (df['Gold.2']*3) + (df['Silver.2']* 2) + (df['Bronze.2'])
    return points
answer_four()
df.head()
```

```python
census_df = pd.read_csv('census.csv')
census_df.head()
```

```python
def answer_five():
    counties = census_df[census_df['SUMLEV']==50]
    return counties.groupby('STNAME').count().COUNTY.idxmax()
answer_five()
```

```python
def answer_six():
    counties = census_df[census_df['SUMLEV']==50]
    threecounties = counties.groupby('STNAME')['CENSUS2010POP'].nlargest(3)
    states = threecounties.groupby('STNAME').sum()
    return list(states.nlargest(3).index)
answer_six()
```

```python
def answer_seven():
    counties = census_df[census_df['SUMLEV']==50]
    population_estimates = ["POPESTIMATE2010","POPESTIMATE2011","POPESTIMATE2012","POPESTIMATE2013","POPESTIMATE2014","POPESTIMATE2015"]
    return counties.iloc[(counties[population_estimates].max(axis=1)-counties[population_estimates].min(axis=1)).argmax()]['CTYNAME']
answer_seven()
```

```python
def answer_eight():
    counties = census_df[census_df['SUMLEV']==50]
    region = counties[(counties['REGION']==1) | (counties['REGION']==2)]
    washington = region[region['CTYNAME'].str.startswith("Washington")]
    popestimate = washington[washington['POPESTIMATE2015'] > washington['POPESTIMATE2014']]
    return popestimate[['STNAME','CTYNAME']]

answer_eight()
```
