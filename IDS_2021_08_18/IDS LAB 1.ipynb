```python
def answer_seven():
    df = df_census.copy()
    df=df[df['SUMLEV'] == 50]
    df = df[['STNAME','CTYNAME','POPESTIMATE2015','POPESTIMATE2014','POPESTIMATE2013','POPESTIMATE2012','POPESTIMATE2011','POPESTIMATE2010']]
    df = df.set_index(['STNAME', 'CTYNAME'])
    df1 = df.apply(lambda df: df.max() - df.min(),axis=1)
    return df1.idxmax()

answer_seven()
```
