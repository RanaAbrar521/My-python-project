****************************** CLEANING AND MAKING OF DATASET ******************************

```python
import pandas as pd
import numpy as np
energy = pd.read_excel ('Energy+Indicators.xls' ,skiprows =17)
```

```python
energy.head()
```

```python
energy.drop(['Unnamed: 0', 'Unnamed: 1'], axis = 1, inplace = True)  #droping 
```

```python
energy.head()
```

```python
energy.rename(	columns={"Unnamed: 2":'Country'}, inplace = True)
energy.rename(	columns={"Petajoules":'Energy Supply'}, inplace = True)
energy.rename(	columns={"Gigajoules":'Energy Supply per Capita'}, inplace = True)
energy.rename(	columns={"%":'% Renewable'}, inplace = True)
```

```python
energy.head()
```

```python
energy.replace('...',np.nan, inplace=True)
```

```python
#now multiply the energy supply column with 1000000 for gigjoule convertion
energy['Energy Supply'] = energy['Energy Supply'] * 1000000
```

```python
energy.head()
```

```python
energy.replace('Republic of Korea','South Korea', inplace = True)
energy.replace(	'United States of America','United States', inplace = True)
energy.replace(	'United Kingdom of Great Britain and Northern Ireland','United Kingdom', inplace = True)
energy.replace(	'China, Hong Kong Special Administrative Region','Hong Kong', inplace = True)

energy.replace('Bolivia (Plurinational State of)','Bolivia', inplace = True)
energy.replace(	'Switzerland17','Switzerland', inplace = True)
```

```python
energy.head()
```

```python
energy['Country'].replace('[^a-zA-Z ]',' ', inplace=True)
```

******************************************* Q10 *******************************************

```python
import warnings
warnings.filterwarnings('ignore')
def answer_ten():
    final_dataset['HighRenew'] = 0
    final_dataset.sort_values('Rank', inplace = True)
    median_renew = int (final_dataset['% Renewable'].median(axis=0))
    for i in range(15):
        if  (final_dataset['% Renewable'].iloc[i]) >= median_renew :
            final_dataset['HighRenew'].iloc[i] = 1
        else:
            final_dataset['HighRenew'].iloc[i] = 0
    return final_dataset['HighRenew']
```

```python
answer_ten()
```

******************************************* Q11 *******************************************

```python
def answer_eleven():
    ContinentDict = {'China':'Asia',
                    'United States':'North America',
                    'Japan':'Asia',
                    'United Kingdom':'Europe',
                    'Russian Federation':'Europe',
                    'Canada':'North America','Germany':'Europe',
                    'India':'Asia',
                    'France':'Europe',
                    'South Korea':'Asia',
                    'Italy':'Europe',
                    'Spain':'Europe',
                    'Iran':'Asia',
                    'Australia':'Australia',
                    'Brazil':'South America'}
    
    country_groups = pd.DataFrame(columns=['size', 'sum', 'mean', 'std'])
    
    for continents, data in final_dataset.groupby(ContinentDict):
        
        size = len(data)
        sum_population_estimate = data['population_estimate'].sum()
        mean_population_estimate = data['population_estimate'].mean()
        std_population_estimate = data['population_estimate'].std()    
        
        country_groups.loc[continents] = [size,sum_population_estimate ,mean_population_estimate ,std_population_estimate ]
        
    return country_groups
```

```python
answer_eleven()
```
