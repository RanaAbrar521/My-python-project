```python
import pandas as pd
import numpy as np
energy = pd.read_excel ('Energy+Indicators.xls' ,skiprows =17)
```

```python
energy.head()
```

```python
energy.drop(['Unnamed: 0', 'Unnamed: 1'], axis = 1, inplace = True)  #droping 
```

```python
energy.head()
```

```python
energy.rename(	columns={"Unnamed: 2":'Country'}, inplace = True)
energy.rename(	columns={"Petajoules":'Energy Supply'}, inplace = True)
energy.rename(	columns={"Gigajoules":'Energy Supply per Capita'}, inplace = True)
energy.rename(	columns={"%":'% Renewable'}, inplace = True)
```

```python
energy.head()
```

```python
energy.replace('...',np.nan, inplace=True)
```

```python
#now multiply the energy supply column with 1000000 for gigjoule convertion
energy['Energy Supply'] = energy['Energy Supply'] * 1000000
```

```python
energy.head()
```

```python
energy.rename(	columns={'Republic of Korea':'South Korea'}, inplace = True)
energy.rename(	columns={'United States of America':'United States'}, inplace = True)
energy.rename(	columns={'United Kingdom of Great Britain and Northern Ireland':'United Kingdom'}, inplace = True)
energy.rename(	columns={'China, Hong Kong Special Administrative Region':'Hong Kong'}, inplace = True)

energy.rename(	columns={'Bolivia (Plurinational State of)':'Bolivia'}, inplace = True)
energy.rename(	columns={'Switzerland17':'Switzerland'}, inplace = True)
```

```python
energy.head()
```

```python
energy['Country'].replace('[^a-zA-Z ]',' ', inplace=True)
```
