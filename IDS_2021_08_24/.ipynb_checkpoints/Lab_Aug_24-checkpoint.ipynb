```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
nobel = pd.read_csv ('nobel.csv')
```

```python
nobel.head()
```

```python
def answer_two():
    sex= nobel['sex'].value_counts().idxmax()
    country = nobel['birth_country'].value_counts().idxmax()
    return sex,country
```

```python
answer_two()
```

```python
# Q3
def answer_three():
    nobel['USA_Born_Winner'] = nobel['birth_country'] == 'United States of America' #making of new column and build a check
    nobel['per_decade'] = (np.floor(nobel['year']/10)*10).astype(int) # making of another new column
    USA_Born_Winner = nobel.groupby(['per_decade'], as_index=False)['USA_Born_Winner'].mean() #calculating the proportions of decades after grouping them

    return USA_Born_Winner
```

```python
answer_three()
```

```python
#Q4
graph = sns.lineplot(x=nobel["per_decade"], y=nobel["USA_Born_Winner"])
graph.yaxis.set_major_formatter(PercentFormatter())
```

```python
#Q5
# Calculating the proportion of female laureates per decade
nobel['female_winner'] = nobel["sex"] == "Female"
prop_female_winners = nobel.groupby(["per_decade", "category"], as_index=False)["female_winner"].mean()

# Plotting USA born winners with % winners on the y-axis
ax = sns.lineplot(x="per_decade", y="female_winner", hue="category", data=prop_female_winners)
```
