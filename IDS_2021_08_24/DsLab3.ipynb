```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
def answer_one():
    data = pd.read_csv('nobel.csv')
    return data.head()
answer_one()
```

```python
def answer_two():
    gender_data =  data['sex'].value_counts()
    
    birth_data = data['birth_country'].value_counts().head()
    
    return gender_data , birth_data

answer_two()
```

```python
def answer_three():
    data['USA Winner'] = data['birth_country'] == 'United States of America'
    
    data['decade'] = (np.floor(data['year']/10)*10).astype(int) 
    
    USA_Winners = data.groupby(['decade'], as_index=False)['USA Winner'].mean()
    
    return USA_Winners

answer_three()
```

```python
def answer_four():
    graph = sns.lineplot(x=data["decade"], y=data["USA Winner"])
    from matplotlib.ticker import PercentFormatter
    graph.yaxis.set_major_formatter(PercentFormatter())

answer_four()
```

```python
def answer_five():
    data['female_winner'] = data['sex']=='Female'
    F_Winners = data.groupby(['decade', 'category'], as_index=False)['female_winner'].mean()
    return F_Winners
answer_five()
```

```python
def answer_six():
    return data[data['sex'] == 'Female'].nsmallest(1,'year')

answer_six()
```

```python
def answer_seven():
    return data.groupby('full_name').filter(lambda x: len(x)>1)

answer_seven()
```

```python
def answer_eleven():
    data['birth_date'] =  pd.to_datetime(data['birth_date'], format='%Y-%m-%d', errors='ignore')
    data['age'] = data['year'] - data['birth_date'].dt.year
    return data.nsmallest(1, 'age')

answer_eleven()
    
```
