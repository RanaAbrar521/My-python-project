```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def answer_one():  
    currency_df = pd.read_csv('coinmarketcap_06122017.csv')
    return currency_df.head(10)
answer_one()
```

```python
def answer_two():
    return currency_df[['id','market_cap_usd']].count()
answer_two()
```

```python
def answer_three():
    desired_df = currency_df[currency_df['market_cap_usd']>0]
    return desired_df[['id','market_cap_usd']].count()
answer_three()
```

```python
def answer_four():
    TopTen = desired_df.nlargest(10,'market_cap_usd')
    TopTen['Market_TopTen_percentage'] = TopTen['market_cap_usd'] / TopTen['market_cap_usd'].sum() * 100
    TopTen.plot.bar(x='id',y='Market_TopTen_percentage')
answer_four()
```

```python
def answer_five():
    COLORS = ['orange', 'green', 'orange', 'cyan', 'cyan', 'blue', 'silver', 'orange', 'red', 'green']
    TopTen.plot.bar(x='id',y='market_cap_usd',log=True,color=COLORS)
answer_five()
```

```python
def answer_six():
    desired_df.plot.line(x='id', y='percent_change_24h',figsize=(10,5))
    desired_df.plot.line(x='id', y='percent_change_7d',figsize=(10,5))
answer_six()
```

```python
def answer_seven():
    TopTen_gain = desired_df.nlargest(10,'percent_change_24h')
    TopTen_gain.plot.bar(x='id', y='percent_change_24h')
    plt.title("Top 10 Biggest Gainers")
    plt.show()

    TopTen_loss = desired_df.nsmallest(10,'percent_change_24h')
    TopTen_loss.plot.bar(x='id', y='percent_change_24h')
    plt.title("Top 10 Biggest Losers")
    plt.show()
    
answer_seven()
```

```python
def answer_eight():
    TopTen_gain = desired_df.nlargest(10,'percent_change_7d')
    TopTen_gain.plot.bar(x='id', y='percent_change_7d')
    plt.title("Top 10 Biggest Gainers")
    plt.show()

    TopTen_loss = desired_df.nsmallest(10,'percent_change_7d')
    TopTen_loss.plot.bar(x='id', y='percent_change_7d')
    plt.title("Top 10 Biggest Losers")
    plt.show()
answer_eight()
```
