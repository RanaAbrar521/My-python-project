```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('coinmarketcap_06122017.csv')
valid_df = df[df['market_cap_usd'] > 0]
valid_df.sort_values(by=['market_cap_usd'], ascending=False, inplace=True)
```

## Q1

```python
def answer_one():
    return df[['id', 'market_cap_usd']].count()
```

```python
answer_one()
```

## Q2

```python
def answer_two():
    return valid_df.head()
```

```python
answer_two()
```

## Q3

```python
def answer_three():
    top_10 = valid_df.head(10)
    sum_all_values = top_10[['market_cap_usd']].sum().iloc[0]
    top_10['market_cap_percentage'] = (top_10['market_cap_usd'] / sum_all_values) * 100
    top_10.plot(kind='bar', x='id', y='market_cap_percentage', figsize=(12,6), grid=True, color="#6f00ff")
    plt.title("Currency Market Cap Percentages")
    plt.xlabel("Crypto Currency")
    plt.ylabel("Percentage")
    plt.show()
#     sns.barplot(data=top_10, x='id', y='market_cap_percentage')
```

```python
answer_three()
```

## Q5

```python
def answer_five():
    valid_df.plot(kind='line', x='id', y='percent_change_24h', figsize=(12,6), grid=True, color="#6f00ff")
    plt.title("Currency Market Cap Change in 24 Hours")
    plt.xlabel("Crypto Currency")
    plt.ylabel("Percentage Change")
    plt.show()
    
    valid_df.plot(kind='line', x='id', y='percent_change_7d', figsize=(12,6), grid=True, color="#6f00ff")
    plt.title("Currency Market Cap Change in 7 Days")
    plt.xlabel("Crypto Currency")
    plt.ylabel("Percentage Change")
    plt.show()
```

```python
answer_five()
```

## Q6

```python
def answer_six():
    
    top_10_gainers = valid_df.nlargest(10, 'percent_change_24h')
    top_10_gainers.plot(kind='bar', x='id', y='percent_change_24h', figsize=(12,6), grid=True, color="#6f00ff")
    plt.title("Currency Market Cap Change in 24 Hours (Gainers)")
    plt.xlabel("Crypto Currency")
    plt.ylabel("Percentage Change")
    plt.show()
    
    top_10_losers = valid_df.nsmallest(10, 'percent_change_24h')
    top_10_losers.plot(kind='bar', x='id', y='percent_change_24h', figsize=(12,6), grid=True, color="#6f00ff")
    plt.title("Currency Market Cap Change in 24 Hours (Losers)")
    plt.xlabel("Crypto Currency")
    plt.ylabel("Percentage Change")
    plt.show()
```

```python
answer_six()
```
