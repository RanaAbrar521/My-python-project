```python
def answer_four():   
    sum_of_topten_market = top_ten[['market_cap_usd']].sum().iloc[0]
    top_ten['market_cap_%'] = (top_ten['market_cap_usd']/sum_of_topten_market)*100
    return top_ten.plot(kind= 'bar', x= 'id',y='market_cap_%')
```

```python
answer_four()
```
