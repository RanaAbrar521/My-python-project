```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def answer_one():
    df = pd.read_csv('disney_movies_total_gross.csv')
    return df

answer_one()
```

```python
def answer_two():
    Top10 = df.sort_values(by = 'inflation_adjusted_gross', ascending = False)
    return Top10.head(10)
answer_two()
```

```python
def answer_three():
    genre = df.groupby(['genre']).mean()
    pop_genre = genre.reset_index()
    pop_genre['inflation_adjusted_gross'] = pop_genre['inflation_adjusted_gross'].astype(int)
    pop_genre = pop_genre.sort_values(by = 'inflation_adjusted_gross', ascending = False)
    return pop_genre.head(10)
answer_three()
```

```python
def answer_four():
    Top10 = answer_two()
    plt.figure(figsize=(10,10))
    sns.barplot(x=pop_genre.genre, y=pop_genre.inflation_adjusted_gross)
    plt.xlabel("Genre")
    plt.ylabel("Gross Income (billions)")
    plt.xticks(rotation = 90)
    plt.tight_layout()
    return plt.show()
answer_four()
```
