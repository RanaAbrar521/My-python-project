```python
import pandas as pd
```

```python
df = pd.read_csv("olympics.csv",index_col=0,skiprows=1)
for col in df.columns:
    if col[:2]== '01':
        df.rename(columns={col:'Gold'+ col[4:]},inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+ col[4:]},inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+ col[4:]},inplace=True)
    if col[:2]=='N^2':
        df.rename(columns={col:'#'+ col[1:]},inplace=True)
        
names_ids = df.index.str.split('\s\(')
df.index = names_ids.str[0]
df['ID'] = names_ids.str[1].str[:3]
df=df.drop('Totals')
df
```

```python
def Question_5():
    counties = df[df['SUMLEV']==50]
    s = counties.groupby('STNAME')['COUNTY'].count().idxmax()
    return s
```

# Q\6

```python
def Question_6():
    counties = df[df['SUMLEV']==50]
    s = counties.groupby('STNAME')['CENSUS2010POP'].nlargest(3)
    x = s.groupby('STNAME').sum()
    return list(x.nlargest(3).index)
```

```python
Question_6()
```

# Q\7

```python
def Question_7():
    counties = df[df['SUMLEV'] == 50]
    s = ["POPESTIMATE2010","POPESTIMATE2011","POPESTIMATE2012","POPESTIMATE2013","POPESTIMATE2014","POPESTIMATE2015"]
    x = counties[s].max(axis=1) - counties[s].min(axis=1)
    return df.loc[x.idxmax()]['CTYNAME']
```

```python
Question_7()
```

# Q\8

```python
def answer_eight():
    counties = df[df['SUMLEV'] == 50]
    s = counties[((counties['REGION']==1) | (counties['REGION']==2)) & counties['CTYNAME'].str.startswith("Washington")]
    x =  s[ s['POPESTIMATE2015'] >  s['POPESTIMATE2014']]
    return  x[['STNAME','CTYNAME']].sort_index()
```
