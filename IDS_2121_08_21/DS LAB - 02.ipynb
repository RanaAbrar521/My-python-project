```python
import pandas as pd
import numpy as np
```

```python
# Energy+Indicators.xls
```

```python
energy = pd.read_excel('Energy+Indicators.xls',skiprows = 17 )
```

```python
energy.head()
```

```python
energy.drop(['Unnamed: 0','Unnamed: 1'],axis = 1,inplace = True)
```

```python
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
```

```python
energy.head()
```

```python
energy.replace('...','NaN',inplace = True)
```

```python
energy.head()
```

```python
energy['Energy Supply'] = energy['Energy Supply']*1000000
```

```python
energy.head()
```

```python
energy.replace('United States of America20','United States', inplace=True)
energy.replace('United Kingdom of Great Britain and Northern Ireland19','United Kingdom', inplace=True)
energy.replace('China, Hong Kong Special Administrative Region3','Hong Kong', inplace=True)
energy.replace('Republic of Korea','South Korea', inplace=True)
energy.replace('Bolivia (Plurinational State of)','Bolivia', inplace=True)
```

```python
energy['Country'] = energy['Country'].str.replace('\d+','')
 
```

```python
energy.head()
```
